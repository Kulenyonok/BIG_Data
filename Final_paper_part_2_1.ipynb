```python
import pandas as pd
```

```python
import re
```

```python
url = 'https://drive.google.com/file/d/1-GrZkR836XfVxuRP-D-epf3ZXJg9xvsR/view?usp=sharing'
```

```python
url = 'https://drive.google.com/uc?id=' + url.split('/')[-2]
```

```python
ratings = pd.read_csv(url, sep = '\t', names = ['user id', 'item id', 'rating', 'timestamp'])
```

```python
ratings.head()
```

```python
url2 = 'https://drive.google.com/file/d/1tw3BAUhCySeJnyREnJw6Ldh4xmO29fcN/view?usp=sharing'
```

```python
url2 = 'https://drive.google.com/uc?id=' + url2.split('/')[-2]
```

```python
names2 = """movie id | movie title | release date | video release date |
  IMDb URL | unknown | Action | Adventure | Animation |
  Childrens | Comedy | Crime | Documentary | Drama | Fantasy |
  Film-Noir | Horror | Musical | Mystery | Romance | Sci-Fi |
  Thriller | War | Western"""
```

```python
names21 = re.split(r'\s+[|]\s*', names2)
names21
```

```python
movies = pd.read_csv(url2, sep = '|', encoding='latin-1', names = names21)
```

```python
movies.head()
```

```python
movies.columns
```

EX_2

```python
ratings_sort = ratings.groupby('user id').count().sort_values(by = 'rating', ascending = False)
ratings_sort
```

```python
super_user = ratings_sort.index[0]
super_user
```

EX_3

```python
ratings_1 = ratings.loc[ratings['user id'] == super_user]
ratings_1
```

EX_4

```python
ratings_1m = ratings_1.merge(movies, how = 'left', left_on='item id', right_on='movie id')
ratings_1m.head()
```

```python
ratings_1m.columns
```

```python
ratings_1m = ratings_1m.drop(columns=['movie id', 'movie title', 'video release date', 'IMDb URL'])
ratings_1m.head()
```

```python
ratings_sort1 = ratings.groupby('item id').count()
ratings_sort1
```

```python
ratings_sort1 = ratings_sort1.drop(columns=['user id', 'timestamp'])
ratings_sort1.head()
```

```python
ratings_sort1.rename(columns={'rating': 'rating_count'}, inplace=True)
ratings_sort1.head()
```

```python
ratings_sort2 = ratings.groupby('item id').sum()
ratings_sort2
```

```python
ratings_sort2 = ratings_sort2.drop(columns=['user id', 'timestamp'])
ratings_sort2.head()
```

```python
ratings_sort2.rename(columns={'rating': 'rating_sum'}, inplace=True)
ratings_sort2.head()
```

```python
ratings_1m = ratings_1m.merge(ratings_sort1, how = 'left', left_on='item id', right_on='item id').merge(ratings_sort2, how = 'left', left_on='item id', right_on='item id')
ratings_1m.head()
```

```python
ratings_1m.columns
```

```python
ratings_1m.dtypes
```

```python
release_year = []
for el in ratings_1m['release date']:
  a = el[-4 :]
  release_year.append(a)
```

```python
ratings_1m['release_year'] =  release_year
ratings_1m.head()
```

```python
ratings_1m.dtypes
```

```python
import numpy as np
```

```python
ratings_1m = ratings_1m.astype({'release_year': np.int64})
```

```python
ratings_1m.dtypes
```

```python
ratings_1m.columns
```

EX_5

```python
X = ratings_1m[['unknown',
       'Action', 'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime',
       'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical',
       'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western',
       'rating_count', 'rating_sum', 'release_year']]
y = ratings_1m['rating']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

EX_6

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

EX_7

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, y_pred)
```
